### 1. **Distribution of Energy Efficiency (SEER2 and EER2)**
   - **Plot:** Histograms or Kernel Density Plots
   - **Benefit:** Visualizes the energy efficiency distribution across products. Helps identify trends in efficiency levels, clusters, or outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check and clean SEER2 and EER2
if 'SEER2 (Btu/Wh)' in df_filtered.columns and 'EER2 (Btu/Wh)' in df_filtered.columns:
    sns.histplot(df_filtered['SEER2 (Btu/Wh)'], kde=True, bins=20, color='blue', label='SEER2')
    plt.title('Distribution of SEER2 (Btu/Wh)')
    plt.xlabel('SEER2 (Btu/Wh)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    sns.histplot(df_filtered['EER2 (Btu/Wh)'], kde=True, bins=20, color='green', label='EER2')
    plt.title('Distribution of EER2 (Btu/Wh)')
    plt.xlabel('EER2 (Btu/Wh)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
else:
    print("SEER2 or EER2 column is missing.")
```

### 2. **Energy Efficiency vs. Cooling Capacity**
   - **Plot:** Scatter Plot (with SEER2 and EER2 on the axes and Cooling Capacity as a third dimension using color or size)
   - **Benefit:** Reveals relationships between energy efficiency and cooling capacity. Useful for identifying if high efficiency correlates with high or low capacity.

```python
plt.scatter(df_filtered['SEER2 (Btu/Wh)'], df_filtered['Cooling Capacity (Btu/h)'], alpha=0.7, label='SEER2', c='blue')
plt.scatter(df_filtered['EER2 (Btu/Wh)'], df_filtered['Cooling Capacity (Btu/h)'], alpha=0.7, label='EER2', c='green')
plt.title('Energy Efficiency vs. Cooling Capacity')
plt.xlabel('Energy Efficiency (Btu/Wh)')
plt.ylabel('Cooling Capacity (Btu/h)')
plt.legend()
plt.show()
```

### 3. **Distribution by Manufacturer Type**
   - **Plot:** Bar Chart
   - **Benefit:** Compares the count of products across different manufacturer types. Useful for understanding market representation.

```python
df_filtered['Manufacturer Type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution by Manufacturer Type')
plt.xlabel('Manufacturer Type')
plt.ylabel('Count')
plt.show()
```

### 4. **Refrigerant Type Distribution**
   - **Plot:** Pie Chart or Bar Chart
   - **Benefit:** Highlights the popularity of refrigerant types and helps evaluate the adoption of low-GWP refrigerants for environmental impact.

```python
df_filtered['Refrigerant Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Refrigerant Type Distribution')
plt.ylabel('')  # Hides the y-label
plt.show()
```

### 5. **Time Trend of Certification**
   - **Plot:** Line Chart (Count of certifications over `Date Certified`)
   - **Benefit:** Shows how certifications have trended over time. Indicates industry activity or market maturity.


```python
df_filtered['Date Certified'] = pd.to_datetime(df_filtered['Date Certified'], errors='coerce')
df.groupby(df['Date Certified'].dt.year).size().plot(kind='line', marker='o', color='orange')
plt.title('Time Trend of Certifications')
plt.xlabel('Year')
plt.ylabel('Number of Certifications')
plt.show()
```

### 6. **Peak Cooling Requirement Compliance**
   - **Plot:** Stacked Bar Chart (Manufacturer Type vs. Compliance)
   - **Benefit:** Examines how manufacturers align with peak cooling requirements.


```python
compliance_counts = df_filtered.groupby('Manufacturer Type')['Meets Peak Cooling Requirements'].value_counts().unstack()
compliance_counts.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Peak Cooling Requirement Compliance by Manufacturer Type')
plt.xlabel('Manufacturer Type')
plt.ylabel('Count')
plt.legend(title='Compliance')
plt.show()
```

### 7. **Tax Credit Eligibility Analysis**
   - **Plot:** Grouped Bar Chart (Tax Credit Eligible vs. Tax Credit Eligible - CAC, National)
   - **Benefit:** Compares products' eligibility for tax credits, highlighting compliance with regional and national incentives.


```python
eligibility_counts = df_filtered[['Tax Credit Eligible', 'Tax Credit Eligible - CAC, National']].apply(pd.Series.value_counts)
eligibility_counts.T.plot(kind='bar', color=['blue', 'green'])
plt.title('Tax Credit Eligibility Analysis')
plt.xlabel('Eligibility Type')
plt.ylabel('Count')
plt.legend(title='Status')
plt.show()
```

### 8. **Series Name vs. Efficiency**
   - **Plot:** Box Plot (Series Name on the x-axis, SEER2/EER2 on the y-axis)
   - **Benefit:** Highlights variation in energy efficiency within product series. Identifies high-performing series.


```python
plt.figure(figsize=(15, 7))
sns.boxplot(x='Series Name', y='SEER2 (Btu/Wh)', data=df_filtered, palette='coolwarm')
plt.title('Energy Efficiency by Series Name')
plt.xticks(rotation=90, ha='right')
plt.xlabel('Series Name')
plt.ylabel('SEER2 (Btu/Wh)')
plt.show()
```

### 9. **Geographic Market Distribution**
   - **Plot:** Horizontal Bar Chart (Markets)
   - **Benefit:** Reveals the distribution of products across different markets. Useful for market segmentation analysis.


```python
market_counts = df_filtered['Markets'].value_counts()
market_counts.plot(kind='barh', color='teal')
plt.title('Market Distribution')
plt.xlabel('Count')
plt.ylabel('Markets')
plt.show()
```

### 10. **Connected Capable Feature**
   - **Plot:** Bar Chart or Pie Chart (Proportion of connected capable products)
   - **Benefit:** Illustrates adoption rates of IoT-enabled features. Useful for understanding smart product penetration.


```python
df_filtered['Connected Capable'].value_counts().plot(kind='bar', color='coral')
plt.title('Connected Capable Product Distribution')
plt.xlabel('Connected Capability')
plt.ylabel('Count')
plt.show()
```

These plots serve multiple purposes:

- **Industry Insights:** Identifying leading manufacturers or brands based on efficiency, certification, or compliance.
- **Consumer Trends:** Understanding what features (e.g., connected capabilities, refrigerant types) are gaining popularity.
- **Policy Impact:** Assessing adherence to energy standards and tax credit eligibility.
- **Market Strategy:** Guiding manufacturers to focus on high-demand or underserved market segments.

```python
from sklearn.preprocessing import LabelEncoder

df_numeric = df_filtered.copy()

le = LabelEncoder()

for i in df_filtered.columns:
    if df_filtered[i].dtype == 'object':
        df_numeric[i] = le.fit_transform(df_numeric[i])

correlation_matrix = df_numeric.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```
